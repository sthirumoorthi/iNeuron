**1. Write a Python Program to Display Fibonacci Sequence Using Recursion?**

```python
def fibonacci(n):
    """
    This is a recursive function which will print the Fibonacci series for n terms.
    """
    
    if n <= 1:
        return n
    else:
        return(fibonacci(n-1) + fibonacci(n-2))
```

```python
nterms = int(input("Enter the number of Fibonacci series terms: "))

# check if the number of terms is valid
if nterms <= 0:
    print("Invalid integer. Plese rerun the program with positive integer")
else:
    print("Fibonacci sequence:")
    for i in range(nterms):
        print(fibonacci(i))
```

**2. Write a Python Program to Find Factorial of Number Using Recursion?**

```python
def factorial(n):
    """
    This is a recursive function which will return the factorial of any input integer
    """
    
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)

    
num = int(input("Enter the number to find its factorial: "))
result = factorial(num)
print(f"The factorial of {num} is: {result}")
```

**3. Write a Python Program to calculate your Body Mass Index?**

```python
height = float(input("Input your height in Meters: "))
weight = float(input("Input your weight in Kilogram: "))

bmi = weight / (height * height)

if ( bmi < 18.5):
    health = "underweight"

elif ( bmi >= 18.5 and bmi < 25):
    health = "Healthy"

elif ( bmi >= 25 and bmi < 30):
    health = "overweight"

elif ( bmi >=30 and bmi < 40):
    health = "obesity"

elif ( bmi >= 40):
    health = "severe obesity"

print(f"Your BMI is: {bmi} and you are: {health}")
```

**5. Write a Python Program for cube sum of first n natural numbers?**

```python
def sumOfCubicNum(n):
    """ 
    This function returns the cubic sum of n natural numbers.
    """
    
    cubic_sum = ( n * (n + 1) / 2 ) ** 2
    return int(cubic_sum)
```

```python
num = int(input(" Enter a number to find its cubic sum: "))
print(f"Cubic sum of first {num} natural number is: {sumOfCubicNum(num)}")
```
