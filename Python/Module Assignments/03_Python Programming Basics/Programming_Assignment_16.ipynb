**Question1.**

    Write a function that stutters a word as if someone is struggling to read it. The first two letters are repeated twice with an ellipsis ... and space after each, and then the word is pronounced with a question mark ?.

**Examples:**

    stutter("incredible") ➞ "in... in... incredible"
    stutter("enthusiastic") ➞ "en... en... enthusiastic"
    stutter("outstanding") ➞ "ou... ou... outstanding"

Hint :- Assume all input is in lower case and at least two characters long.

```python
def stutter(word):
    """
    This function will stutters the given word. 
    Repeat first two letters twice with an ellipsis and space after each then the word is pronounced with '?'
    """
    s = word[:2]
    return (2 * (s + '... ')) + word + '?'
```

```python
try:
    my_str = input("  Enter any word: ")
except Exception as err:
    print(err)
else:
    print("\n Result is:")
    print(stutter(my_str))
finally:
    print("\n ***** END *****")
```

**Question 2.**

    Create a function that takes an angle in radians and returns the corresponding angle in degrees rounded to one decimal place.

**Examples:**

    radians_to_degrees(1) ➞ 57.3
    radians_to_degrees(20) ➞ 1145.9
    radians_to_degrees(50) ➞ 2864.8

```python
def radians_to_degrees(radian):
    """
    This function will convert the input radians and converts it to degree with one decimal.
    """
    pi = 3.14159
    degree = radian * (180/pi)
    
    return degree

# Main driver
try:
    radian = int(input("  Enter any number (radian): "))
    result = radians_to_degrees(radian)
except Exception as err:
    print(err)
else:
    print(f"Corresponding degree is : {result:.1f}")
finally:
    print("\n ***** END *****")
```

**Question 3.**

    In this challenge, establish if a given integer num is a Curzon number. If 1 plus 2 elevated to num is exactly divisible by 1 plus 2 multiplied by num, then num is a Curzon number.

    Given a non-negative integer num, implement a function that returns True if num is a Curzon number, or False otherwise.

**Examples:**

    is_curzon(5) ➞ True
     2 ** 5 + 1 = 33
     2 * 5 + 1 = 11
     33 is a multiple of 11
    is_curzon(10) ➞ False
     2 ** 10 + 1 = 1025
     2 * 10 + 1 = 21
     1025 is not a multiple of 21
    is_curzon(14) ➞ True
     2 ** 14 + 1 = 16385
     2 * 14 + 1 = 29
     16385 is a multiple of 29

```python
def is_curzon(my_num):
    """
    This function will check if the given input integer is a curzon number or not.
    curzon number = (2^n + 1) should be multiple of (2 * n + 1)
    """
    power = pow(2, my_num) + 1
    product = (2 * my_num + 1)
    
    if (power % product == 0):
        return True
    
    return False
```

```python
# Main driver
try:
    my_num = int(input("  Enter any number: "))
    if (my_num < 0):
        raise ValueError("Please rerun the program with non-negative number")
    else:
        result = is_curzon(my_num)

except ValueError as er:
    print(er)
except Exception as err:
    print(err)
    
else:
    if result:
        print(f"{result}. \nGiven number is a curzon number")
    else:
        print(f"{result}. \nGiven number is not a curzon number")
finally:
    print("\n ***** END *****")
```

**Question 4.**

    Given the side length x find the area of a hexagon.




**Examples:**

    area_of_hexagon(1) ➞ 2.6
    area_of_hexagon(2) ➞ 10.4
    area_of_hexagon(3) ➞ 23.4

```python
import math

def area_of_hexagon(my_num):
    """
    This function will return the area of Hexagon for the given integer.
    Area of hexagon = (3 √3(n)2 ) / 2
    """
    
    return ((3 * math.sqrt(3) * (my_num * my_num)) / 2)
```

```python
# Main driver
try:
    my_num = int(input("  Enter the side for the hexagon: "))
    if (my_num < 0):
        raise ValueError("Please rerun the program with non-negative number")
    else:
        result = area_of_hexagon(my_num)

except ValueError as er:
    print(er)
except Exception as err:
    print(err)
    
else:
    print(f"\nArea of Hexagon is: {round(result, 1)}")
finally:
    print("\n ***** END *****")
```

**Question 5.**

    Create a function that returns a base-2 (binary) representation of a base-10 (decimal) string number. To convert is simple: ((2) means base-2 and (10) means base-10) 
    010101001(2) = 1 + 8 + 32 + 128.

    Going from right to left, the value of the most right bit is 1, now from that every bit to the left will be x2 the value, value of an 8 bit binary numbers are (256, 128, 64, 32, 16, 8, 4, 2, 1). 

**Examples:**

    binary(1) ➞ "1"
     1*1 = 1
    binary(5) ➞ "101"
     1*1 + 1*4 = 5
    binary(10) ➞ "1010"
     1*2 + 1*8 = 10

```python
def decimalToBinary(num):
    """
    This function will convert the decimal number to a binary form.
    """
     
    binary_number = 0
    count = 0
    while (num != 0):
        rem = num % 2
        c = pow(10, count)
        binary_number += rem * c
        num //= 2
         
        # count to store exponent value
        count += 1
     
    return binary_number
```

```python
# Main driver
try:
    my_num = int(input("  Enter the number in decimal format: "))
    if (my_num < 0):
        raise ValueError("Please rerun the program with non-negative number")
    else:
        result = decimalToBinary(my_num)

except ValueError as er:
    print(er)
except Exception as err:
    print(err)
    
else:
    print(f"\nBinary form is: {result}")
finally:
    print("\n ***** END *****")
```
